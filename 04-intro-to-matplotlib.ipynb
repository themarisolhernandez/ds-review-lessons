# 04 - Introduction to Matplotlib
**Matplotlib** is a comprehensive library used for creating static, animated, and interactive visualizations in Python. It is a widely-used tool in data science to plot data in a visual format that allows us to better understand patterns, trends, and outliers. It can generate plots like line graphs, bar charts, scatter plots, histograms, and more.

## Key Features of Matplotlib
- **Ease of Use**: Easy syntax for creating basic visualizations.

- **Customization**: Highly customizable, enabling control over plot elements like color, line styles, and legends.

- **Integration**: Integrates seamlessly with libraries like NumPy and Pandas.

## Installation

Once installed, you can import it as follows:

```python
import matplotlib.pyplot as plt
```

Matplotlib is commonly imported with the alias `plt`. You only need to import a library **once** in your notebook to be able to use it in subsequent cells.

## Basic Line Plot
A **line plot** is the simplest way to visualize data. The line plot represents information at points connected by lines. It is useful to show the evolution of one or more data series along an axis, typically time. Hereâ€™s how you can create one:

```python
# Data to plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a line plot
plt.plot(x, y)

# Add a title, axis labels
plt.title('Basic Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Show the plot
plt.show()
```

## Scatter Plot
**Scatter plots** are useful for showing the relationship between two variables

```python
# Create data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Create a scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Basic Scatter Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Show the plot
plt.show()
```

## Bar Plot
**Bar plots** are useful for comparing categories.

```python
# Data to plot
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Create a bar plot
plt.bar(categories, values)

# Add a title and axis labels
plt.title('Basic Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

## Histogram
**Histograms** are used to represent the distribution of data.

```python
import numpy as np

# Generate random data
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30)

# Add a title and labels
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## Customizing Plots
Matplotlib allows you to customize plots in numerous ways. You can change line styles, colors, markers, and more. Letâ€™s explore a few options:

```python
# Data to plot
x = [0, 1, 2, 3]
y1 = [0, 1, 4, 9]
y2 = [0, 1, 2, 3]

# Plot two lines with different styles
plt.plot(x, y1, label='y = x^2', color='blue', marker='o', linestyle='--')
plt.plot(x, y2, label='y = x', color='green', marker='x', linestyle='-.')

# Add labels, title, and legend
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Customized Plot')
plt.legend()

# Show the plot
plt.show()
```

## Subplots
Sometimes you may want to display multiple plots side-by-side. This can be done using subplots.

```python
# Create subplots
fig, axs = plt.subplots(2, 2)

# Plot in different subplots
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Plot 1')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Plot 2')

axs[1, 0].scatter(x, y1)
axs[1, 0].set_title('Plot 3')

axs[1, 1].bar(x, y2)
axs[1, 1].set_title('Plot 4')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

## Saving and Exporting Plots
You can save your plot as an image using the `savefig()` function.

```python
# Save the plot
plt.plot(x, y1)
plt.title('Saving Example')
plt.savefig('imgs/my_plot.png')

# Display the plot
plt.show()
```
